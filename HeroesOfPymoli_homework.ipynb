```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Drop duplicate SN's
purchase_data_no_dup = purchase_data.drop_duplicates(["SN"])
#Get the value count for SN
total_players = purchase_data_no_dup["SN"].value_counts()
#Get the value of the total players
total_players_1 = total_players.sum()
#Place the total players amount into a new Data Frame
total_players_table = pd.DataFrame({"Total Players": [total_players_1]})
total_players_table
```

```python
#Get the unique number of items
unique_items = len(purchase_data['Item Name'].unique())
#Get the average price
average_price = purchase_data['Price'].mean()
#Get the total number of purchases
total_purchases = purchase_data['Item ID'].count()
#get the total revenue
total_rev = purchase_data['Price'].sum()
#Put everything into one table
summary_table = pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": [average_price],
                              "Number of Purchases": [total_purchases],
                              "Total Revenue": [total_rev]})
#format table
summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:,.2f}".format)
summary_table
```

```python
grouped_gender = purchase_data.groupby(['Gender'])
#grouped_gender['Total Count'] = grouped_gender.count()
total_count = grouped_gender.count()
total_count['Total Count'] = total_count['Purchase ID']
#sum1 = total_count.sum() to get total amount of players
total_count['Percentage of Players'] = (total_count['Total Count'] / 780) * 100
total_count1 = total_count[['Total Count','Percentage of Players']]
total_count1['Percentage of Players'] = total_count1['Percentage of Players'].map("{:.2f}".format)
total_count1
```

```python
sum2 = grouped_gender.sum()
total_count1['Total Purchase Value'] = sum2['Price']
avg = grouped_gender.mean()
total_count1['Average Purchase Price'] = avg['Price']
total_count1['Avg Total Purchase per Person'] = total_count1['Total Purchase Value'] / total_count1['Total Count']
analysis = total_count1.rename(columns={"Total Count":"Purchase Count"})
analysis1 = analysis[['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]
analysis1['Average Purchase Price'] = total_count1['Average Purchase Price'].map("${:.2f}".format)
analysis1['Total Purchase Value'] = total_count1['Total Purchase Value'].map("${:.2f}".format)
analysis1['Avg Total Purchase per Person'] = total_count1['Avg Total Purchase per Person'].map("${:.2f}".format)
analysis1
```
